```python
import pandas as pd
```

```python
file = "nuclearincidents.xlsx"
```

```python
df = pd.read_excel(file)
df
```

```python
# Function to extract latitude and longitude from the Location column
def extract_lat_lon(location):
    try:
        if isinstance(location, str):
            # Handle the different formats in the Location data
            if "°" in location:
                # Degrees, minutes, seconds format
                parts = location.split("°")
                lat_part = parts[0].strip()
                lon_part = parts[1].split(",")[1].strip().replace("E", "").replace("W", "-").replace("S", "-")
                lat = float(lat_part)
                lon = float(lon_part)
            elif "Lat:" in location and "Lon:" in location:
                # Lat/Lon format
                lat_part = location.split("Lat:")[1].split("Lon:")[0].strip().replace("N", "").replace("S", "-")
                lon_part = location.split("Lon:")[1].strip().replace("E", "").replace("W", "-")
                lat = float(lat_part)
                lon = float(lon_part)
            else:
                # Decimal format
                lat, lon = map(float, location.split(","))
            return lat, lon
    except:
        return None, None
    return None, None

# Applying the function to extract Latitude and Longitude
df['Latitude'], df['Longitude'] = zip(*df['Location'].apply(extract_lat_lon))

# Dropping the original Location column
df.drop(columns=['Location'], inplace=True)
```
